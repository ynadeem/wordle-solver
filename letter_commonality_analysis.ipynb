```python
import pandas as pd
import numpy as np
from string import ascii_uppercase
from collections import Counter
```

```python
df = pd.read_csv('all_word_options.csv', header=None, names=['all_words'])
df.drop_duplicates(inplace=True)
print(f'Data read successful, found {df.shape[0]:,} unique words')
```

```python
df.head()
```

```python
df = df[['all_words']].join(pd.DataFrame([Counter(word) for word in df['all_words'].str.upper()]).reindex(list(ascii_uppercase), axis=1).fillna(0).astype(int))
```

```python
df.head()
```

```python
df_counts_once = df.replace([2,3,4,5],1)
df_counts_once.head()
```

```python
summary_df_counts_all = df[list(ascii_uppercase)].sum().to_frame(name='count_of_letter')
summary_df_counts_all['pct_count'] = summary_df_counts_all['count_of_letter'] / summary_df_counts_all['count_of_letter'].sum()
summary_df_counts_all
```

```python
summary_df_counts_once = df_counts_once[list(ascii_uppercase)].sum().to_frame(name='count_of_letter')
summary_df_counts_once['pct_count'] = summary_df_counts_once['count_of_letter'] / summary_df_counts_once['count_of_letter'].sum()
summary_df_counts_once
```

```python
summary_df_counts_all.sort_values(by='pct_count', ascending=False)
```

```python
summary_df_counts_once.sort_values(by='pct_count', ascending=False)
```
