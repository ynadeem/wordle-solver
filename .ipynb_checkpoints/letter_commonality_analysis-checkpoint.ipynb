```python
%load_ext lab_black
import pandas as pd
import numpy as np
from string import ascii_uppercase
from collections import Counter
```

```python
df_wordle_options = pd.read_csv('all_word_options.csv', header=None, names=['all_words'])
df_wordle_options.drop_duplicates(inplace=True)
print(f'Data read successful, found {df_wordle_options.shape[0]:,} unique words')
```

```python
df_wordle_options.head()
```

```python
df_wordle_options = df_wordle_options[['all_words']].join(pd.DataFrame([Counter(word) for word in df_wordle_options['all_words'].str.upper()]).reindex(list(ascii_uppercase), axis=1).fillna(0).astype(int))
```

```python
df_wordle_options.head()
```

```python
df_counts_once = df_wordle_options.replace([2,3,4,5],1)
df_counts_once.head()
```

```python
summary_df_counts_all = df_wordle_options[list(ascii_uppercase)].sum().to_frame(name='count_of_letter')
summary_df_counts_all['pct_count'] = summary_df_counts_all['count_of_letter'] / summary_df_counts_all['count_of_letter'].sum()
summary_df_counts_all
```

```python
summary_df_counts_once = df_counts_once[list(ascii_uppercase)].sum().to_frame(name='count_of_letter')
summary_df_counts_once['pct_count'] = summary_df_counts_once['count_of_letter'] / summary_df_counts_once['count_of_letter'].sum()
summary_df_counts_once
```

```python
summary_df_counts_all.sort_values(by='pct_count', ascending=False)
```

```python
summary_df_counts_once.sort_values(by='pct_count', ascending=False)
```
